## Geolocalização das Escolas e Unidades de Saúde em Natal
Análise de dados de unidades de Escolas e Unidades de Saúde em Natal



### Quais informações estão sendos mostradas:

- Mapa com as localizações das Escolas em Natal

- Mapa de calor em relação ao número de Docentes em cada Escola

- Gráfico de barras, comparando o número de Escolas por Regiões

- Mapa com as localizações das Unidades de Saúde em Natal

- Gráfico de barra, comparando o número de Funcionários por Distritos


```python
# Load relevant libraries
import pandas as pd
import numpy as np
import matplotlib
import folium
import geocoder
from tqdm import tqdm
```

```python
data = pd.read_csv('escolas_por_regioes_administrativas.csv', encoding = 'utf_8',sep = ';')
data2 = pd.read_csv('estabelecimento.csv', encoding = 'utf_8',sep = ';')
data3 = pd.read_csv('total.csv', encoding = 'utf_8',sep = ';')
data4 = pd.read_csv('unidades_de_saude.csv', encoding = 'utf_8',sep = ';')
data5 = pd.read_csv('profissionais_das_unidades.csv', encoding = 'utf_8',sep = ';')
```

```python
data.head()
```

```python
data['lat'], data['long'] = [0, 0]
data['geocode_input'] = ""
data['count'] = 0
data.head()
```

**Relacionando a Latitude e Longitude de cada uma das Escolas através do CEP.**

```python
for i in tqdm(range(len(data))):
    g = geocoder.google(data.loc[i,'CEP'])
    data.ix[i,'lat'] = g.lat 
    data.ix[i,'long'] = g.lng
    data.ix[i,'geocode_input'] = g.address
print('Geocoding complete!')
```

```python
data.head()
```

```python
data2['count'] = 0
data2['count'] = data2['Ens. Fundamental'] + data2['Ens. Médio Magistério'] + data2['Ens. Médio'] + data2['Ens. Superio'] + data2['Especialização'] + data2['Mestrado'] + data2['Doutorado'] + data2['Nenhum']
```

```python
data2.head()
```

**Fazendo a contagem de cada um dos Docentes, com as escolas em que cada um deles trabalha.**


```python
for i in tqdm(range(len(data))):
    for j in tqdm(range(len(data2))):
        if(data.ix[i, 'ESTABELECIMENTO'] == data2.ix[j, 'Estabelecimento']):
            data.ix[i,'count'] = data2.ix[j, 'count']
```

```python
data.head()
```

```python
mapc = [0, 30]
zoom = 2
# Create map object
map_marker = folium.Map(location=mapc, zoom_start=zoom)
map_heat = folium.Map(location=mapc, zoom_start=zoom )

```

**Marcando cada uma das posições das escolas no mapa com seu respectivo nome.**

```python
for j in tqdm(range(len(data))):
    if all(~np.isnan([data.ix[j,'lat'], data.ix[j,'long'], data.ix[j,'count']])):   
        folium.Marker([data.ix[j,'lat'], data.ix[j,'long']], popup = data.ix[j,'ESTABELECIMENTO'],
        icon=folium.Icon(color='blue',icon='info-sign')
        ).add_to(map_marker)
        
map_marker
```

**Criando um Mapa de calor baseado no número de funcionários de cada uma das escolas.**

```python
from folium.plugins import HeatMap

coordinates = []

for j in tqdm(range(len(data))):
    if all(~np.isnan([data.ix[j,'lat'], data.ix[j,'long'], data.ix[j,'count']])):
        coordinates.append([data.ix[j,'lat'], data.ix[j,'long'], data.ix[j,'count']])

# Show the map
HeatMap(coordinates).add_to(map_heat)
map_heat
```

# Escolas por Regiões

Gráfico de barras, comparando o número de Escolas por Regiões

### Análise
   No gráfico abaixo, podemos comparar cada uma das regiões de Natal e a quantidade de escolas presentes em cada uma delas.
   Analisando o gráfico, vemos que a Região Norte possui o maior número de escolas com um total de 62 unidades. Em seguida temos a Região Oeste com cerca de 45 unidades e por último as Regiões Sul e Leste com o menor número, com  cerca de 19 unidades em cada.
   

```python
TOOLS = 'box_zoom,box_select,crosshair,resize,reset,wheel_zoom, pan'

p = Bar(data3, 'Região Administrativa', values='Total',color = 'Região Administrativa', xlabel="Região Administrativa", 
        ylabel = "Total", title="Nº de Escolas por Regiões (Natal)", 
        legend = 'top_left',  tools = TOOLS)

output_notebook()

# Display the plot
show(p)
```
