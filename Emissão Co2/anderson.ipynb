```python

# Import output_notebook and show from bokeh.io
from bokeh.io import output_notebook, show

# Import row from bokeh.layouts
from bokeh.layouts import layout, row, gridplot,column
from bokeh.sampledata.autompg import autompg as df
from bokeh.models.widgets import Panel,Tabs


# Import figure from bokeh.plotting
from bokeh.plotting import figure

from bokeh.charts import Histogram, Bar, output_file, Scatter
from bokeh.sampledata.autompg import autompg as df

# Import the ColumnDataSource class from bokeh.plotting
from bokeh.plotting import ColumnDataSource

from bokeh.models.widgets import Panel,Tabs

# Import pandas as pd
import pandas as pd

import numpy as np


# Import the fertility.csv data: data
data = pd.read_csv("CO2_passenger_cars_v1222.csv", sep = ';', encoding = 'latin2')

TOOLS = 'box_zoom,box_select,crosshair,resize,reset,wheel_zoom, pan'
```

```python


Paises = Bar(data, 'MS', values='ec (cm3)',color = 'MS', xlabel="Países", 
        ylabel = "Emissão (Cm³)", title="Emissão de CO2 por paises", 
        legend = '',  tools = TOOLS)



output_notebook()
#output_file("bar2.html")

# Display the plot
show(Paises)
```
