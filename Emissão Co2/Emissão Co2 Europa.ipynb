# Gráfico 1

Gráfico de comparação dos países pelo nível de emissão de CO2 em cada um deles

### Análise
   No gráfico abaixo, podemos comparar cada um dos países pertecentes a União Européia no ano de 2015 e analisar qual deles possui um elevado índice de emissão de CO2 em seus veículos.
   Com esse gráfico de barras, podemos ver claramente no eixo X, a listagem de todos os países pelo eixo Y, os níveis de CO2 em centímetros cúbicos (cm³).
   Olhando rapidamente, vemos que a Dinamarca possui o maior índice de emissão de CO2 pelos veículos que lá trafegam com um valor pouco maior que 1.200e+8 seguido da Áustria com valor em torno de 1.000e+8.
   Os Países menos poluentes são o Chipre, Letônia e a Bélgica, quase sem índices de poluição.



```python
# Import figure from bokeh.plotting
from bokeh.plotting import figure

# Import output_notebook and show from bokeh.io
from bokeh.io import output_notebook, show

# Import row from bokeh.layouts
from bokeh.layouts import layout, row, gridplot,column

from bokeh.charts import Histogram, Bar, output_file, Scatter
from bokeh.sampledata.autompg import autompg as df

# Import the ColumnDataSource class from bokeh.plotting
from bokeh.plotting import ColumnDataSource

from bokeh.models.widgets import Panel,Tabs

# Import pandas as pd
import pandas as pd

import numpy as np

# Import the fertility.csv data: data
#data = pd.read_csv("CO2_passenger_cars_v1222.csv", sep = None)
data = pd.read_csv("CO2_passenger_cars_v1222.csv", sep = ';', encoding = 'latin2')



TOOLS = 'box_zoom,box_select,crosshair,resize,reset,wheel_zoom, pan'

Paises = Bar(data, 'MS', values='ec (cm3)',color = 'MS', xlabel="Países", 
        ylabel = "Emissão (Cm³)", title="Emissão de CO2 por paises", 
        legend = '',  tools = TOOLS)



output_notebook()
#output_file("bar2.html")

# Display the plot
show(Paises)
```

# Gráfico 3

Gráfico dos combustíveis pelo peso dos veículos

### Análise
   No gráfico abaixo, podemos comparar cada um dos combustíveis e ao peso de cada um dos veículos.
  Nele, temos um gráfico de dispersão, onde podemos ver que mesmo o carro tendo um peso mediano (1000kg à 2000kg), ele possui uma poluição maior do que carros mais pesados (>=5000kg).
   Logo, já que o gráfico converge mais para os pesos medianos, do que para os carros mais pesados, podemos concluir que o peso não necessariamente influencia na poluição dos veículos, pois quanto maior o peso a poluição deveria ser maior, que não é o caso.
   

```python
#data2 = data.replace(np.nan,0.0, regex=True)

s = data[['ec (cm3)']].groupby(data['m (kg)']).sum()

y = s.values.tolist()

x = s.index.tolist()

peso = figure(x_axis_label='Peso', y_axis_label='Emissão de Co2 (cm³)', tools = TOOLS)

peso.circle(x, y, color = 'navy')

output_notebook()
show(peso)
```
